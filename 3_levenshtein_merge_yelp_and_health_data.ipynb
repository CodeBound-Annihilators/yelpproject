## Merge Health DataFrame with Yelp DataFrame

### This Merging on something as common as a 3 digit address number will lead to a lot of correct matches, along with a lot of incorrect matches.  Levenshtein Distances will then be used to filter out the incorrect ones.

```python
merged = health.merge(yelp, on='house_numbers', how='left')[['inspection_date', 'inspector', 'number_of_violations', 'violations', 'zip_code', 'house_numbers', 'categories', 'transactions', 'distance', 'review_count', 'longitude', 'latitude', 'name', 'sector', 'district', 'total_score', 'rating', 'price', 'establishment_x', 'establishment_y', 'address_x', 'address_y', 'trimmed_name_x', 'trimmed_name_y', 'clean_address_x',  'clean_address_y']]
merged.dropna(inplace=True)
```

```python
# merged = health.merge(yelp, on='house_numbers', how='left')[['establishment_x','establishment_y','address_x','address_y', 'trimmed_name_x','trimmed_name_y', 'clean_address_x',  'clean_address_y']]
# merged.dropna(inplace=True)
```

```python
merged = merged.reset_index()
```

### Replace street incorrect street names with correct ones

```python
merged['clean_address_x'] = merged['clean_address_x'].str.replace('ih10', 'interstate 10')
merged['clean_address_x'] = merged['clean_address_x'].str.replace('fredsbg', 'fredricksburg')
merged['clean_address_x'] = merged['clean_address_x'].str.replace('loop 410w', 'nw lp 410')
merged['clean_address_x'] = merged['clean_address_x'].str.replace('military hyw', 'military hwy')
merged['clean_address_x'] = merged['clean_address_x'].str.replace('nw ', '')
merged['clean_address_x'] = merged['clean_address_x'].str.replace('san pedroe', 'san pedroe')
merged['clean_address_x'] = merged['clean_address_x'].str.replace('loop 410e', 'ne loop 410')
merged['clean_address_x'] = merged['clean_address_x'].str.replace('military hy', 'military hwy')
```

### Import Levenshtein library in order to filter out names and address numbers that don't match

```python
from Levenshtein import distance
```

```python
distance('bourb street seafood kitchen', 'amigos snacks more')
```

### Create a 'levenshtein distance' between the clean_address_x and clean_address_y columns

```python
merge_names = merged[['clean_address_x', 'clean_address_y']]

s = []

for i in range(len(merged)):
    p = distance(str(merge_names['clean_address_x'][i]), str(merge_names['clean_address_y'][i]))
    print(p)
    s.append(p)
    

    
```

```python
# merge_names['distance'] = d
merged['address_distance'] = s
print(merge_names)
```

### Create a 'levenshtein distance' between the trimmed_name_x and trimmed_name_y columns

```python
merge_names = merged[['trimmed_name_x', 'trimmed_name_y']]

d = []

for i in range(len(merged)):
    p = distance(str(merge_names['trimmed_name_x'][i]), str(merge_names['trimmed_name_y'][i]))
    print(p)
    d.append(p)
    

    
    
# merge_names['distance'] = d
merged['name_distance'] = d
print(merge_names)
```

## Use levenshtein distance values to pull out the matching address values

### This levenshtein distance is basically being used as a filter.  In this case, we used two distances, one between the restaurant name, and one between the address name.  By using multiple two different levenshtein distances it reduces the odds that a 'bad' match will make it through.  

```python
merged = merged[(merged.name_distance < 7) & (merged.address_distance < 4)].sort_values(by='address_distance', ascending=False)
```
