```python
import pandas as pd
# import re
# from pandas.io.json import json_normalize
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

## Import Yelp API json responses

```python
import json
with open('1_17_response.json') as f:
  responses = json.load(f)
```

## Create columns variable

```python
columns = ['alias', 'categories', 'coordinates', 'display_phone', 'id', 'image_url', 'is_closed', 'location', 'name', 'phone', 'price', 'rating', 'review_count', 'transactions', 'url']

```

## Create DataFrame

```python
df = pd.DataFrame(columns=columns)
```

## Loop through 'response' to grab all businesses and their information

```python
for response in range(227):
#     print(responses[response]['businesses'])
    df1 = pd.DataFrame(responses[response]['businesses'])
    df = df.append(df1)
```

```python
df
```

## Reset Index

```python
df = df.reset_index().drop(columns='index')
```

## Garrett will explain this

```python
c_list = {}
check_list= []
for r in responses:
#     print('!!!!')
    for rr in r['businesses']:
        #print(rr)
        print('<<<>>>')
        check_list.append(rr['name'])
        
        if rr['name'] not in list(c_list.keys()):
            c_list.update({rr['name'] : []})
            for c in rr['categories']:
                print(c['title'])
                c_list[rr['name']].append(c['title'])
```

## Loops through list and pulls out restaurant categories

```python
restaurant_list = []
for restaurants in c_list:
    print(c_list[restaurants])
    restaurant_list.append(c_list[restaurants])
```

## Create 'data' dataframe for restaurant categories

```python
data = pd.DataFrame()
data['names'] = check_list
```

## Create categoreos column 

```python
data['categoreos'] = data['names'].apply(lambda x: c_list[x])
```

## Join 'df' on 'data'

```python
df = df.join(data)
```

## Seperating Address information into number, street name, and zipcode

```python
#responses[0]['businesses'][3]['coordinates']['latitude']
addresses = []
for r in responses:
#     print('!!!!')
    #for ii in range(40):
    for rr in r['businesses']:
#         print(rr['coordinates']['latitude'])
        result = rr['location']['address1']
        addresses.append(result)
```

```python
df = df.reset_index().drop(columns='index')
```

```python
#responses[0]['businesses'][3]['coordinates']['latitude']
longitudes = []
for r in responses:
#     print('!!!!')
    #for ii in range(40):
    for rr in r['businesses']:
#         print(rr['coordinates']['latitude'])
        result = rr['coordinates']['longitude']
        longitudes.append(result)
```

## Prepare address by stripping and lowercasing it

```python
df.address = df.address.str.strip().str.lower()
```

```python
df.address = df.address.fillna(np.nan)
```

```python
df.address.dropna(inplace=True)
```

```python
df.address[df.address.isna()]
```

```python
df = df.reset_index().drop(columns='index')
```

## Split strings in address into a list of strings

```python
df['seperated_address'] = df.address.apply(lambda x: str(x).split())
```

## Split numbers off address and place in own column

```python
df['house_numbers'] = df.seperated_address.apply(lambda x: str(x[0]) if len(x)>0 else np.nan)
```

## drop NaNs

```python
df.house_numbers.dropna(inplace=True)
```

## Split street names off address

```python
df['split_address'] = df.address.apply(lambda x: str(x).split()[1:])
```

## Change 'name' to 'establishment' in preparation for merging DataFrames

```python
df['establishment'] = df['name'] 
```

## Trim all extraneous info off of street names

```python

df['trimmed_name'] = df.establishment
df['trimmed_name'] = df['trimmed_name'].str.strip()
df['trimmed_name'] = df['trimmed_name'].str.lower()
df['trimmed_name'] = df['trimmed_name'].str.replace('the ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('on ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('in ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('an ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('of ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('in ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('a ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('el ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('la ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('las ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('los ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('by ', ' ')
df['trimmed_name'] = df.trimmed_name.str.replace('de ', ' ')
```

```python
df['split_address'] = df.split_address.str.join(" ")
df['clean_address'] = df['split_address'].str.replace(' st', '')
df['clean_address'] = df['clean_address'].str.replace(' dr', '')
df['clean_address'] = df['clean_address'].str.replace(' rd', '')
df['clean_address'] = df['clean_address'].str.replace(' av', '')
df['clean_address'] = df['clean_address'].str.replace(' blvd', '')
df['clean_address'] = df['clean_address'].str.replace(' ct', '')
df['clean_address'] = df['clean_address'].str.replace(' n', '')
df['clean_address'] = df['clean_address'].str.replace(' s', '')
df['clean_address'] = df['clean_address'].str.replace(' e', '')
df['clean_address'] = df['clean_address'].str.replace(' w', '')
df['clean_address'] = df['clean_address'].str.replace(' bv', '')
```

```python
# df['trimmed_name'] = df['trimmed_name'].str.lower()
# df['trimmed_name'] = df['trimmed_name'].str.strip()
df['trimmed_name'] = df['trimmed_name'].str.replace("'", '')
df['trimmed_name'] = df['trimmed_name'].str.replace("-", '')
df['trimmed_name'] = df['trimmed_name'].str.replace("&", '')
df['trimmed_name'] = df['trimmed_name'].str.replace("#", '')
```
