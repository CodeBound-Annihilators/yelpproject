```python
import pandas as pd
import numpy as np
```

```python
inspectors = pd.read_csv('health_inspection.csv')
```

```python
inspectors = inspectors.drop(columns='Unnamed: 0')
```

```python
inspectors.columns = ['address', 'district', 'complete_address', 'establishment', 'inspection_date', 'inspector', 'link', 'sector', 'total_score', 'violations']

```

```python
inspectors.address = inspectors.address.str.lower().str.strip()
```

```python
inspectors.establishment = inspectors.establishment.str.lower().str.strip()
```

```python
inspectors
```

```python
inspectors['separated_address'] = inspectors.address.apply(lambda x: x.split())
```

```python
inspectors['house_numbers'] = inspectors.separated_address.apply(lambda x: x[0])
```

```python
inspectors['split_address'] = inspectors.address.apply(lambda x: x.split()[1:])
```

```python
inspectors.head()
```

```python
inspectors['trimmed_name'] = inspectors.establishment.str.replace('the ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('on ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('in ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('an ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('of ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('in ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('a ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('el ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('la ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('las ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('los ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('by ', ' ')
inspectors['trimmed_name'] = inspectors.trimmed_name.str.replace('de ', ' ')
```

```python
inspectors['split_address'] = inspectors.split_address.str.join(" ")
inspectors['clean_address'] = inspectors['split_address'].str.replace(' st', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' dr', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' rd', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' av', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' blvd', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' ct', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' n', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' s', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' e', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' w', '')
inspectors['clean_address'] = inspectors['clean_address'].str.replace(' bv', '')
```
